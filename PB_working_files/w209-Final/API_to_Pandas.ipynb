```python
import requests
import json
import pandas as pd
```

```python
# Search parameters - API limits request to 10 years and 25 series at a time.
series_guide1=pd.read_csv('Series_guide_BLS1.csv',header = 0)
series_guide2=pd.read_csv('Series_guide_BLS2.csv',header = 0) 
series_guide3=pd.read_csv('Series_guide_BLS3.csv',header = 0) 


series_ids1 = series_guide1['Series_ID'].to_list()
series_ids2 = series_guide2['Series_ID'].to_list()
series_ids3 = series_guide3['Series_ID'].to_list()

print(series_ids1)
print(series_ids2)
print(series_ids3)

regkey = 'a8a89503e0a94730886a7f1fde1d55fa'

start_year = 2018
end_year = 2022
```

```python
# Build message to send to API.
headers = {'Content-type': 'application/json'}
data1 = json.dumps({"seriesid": series_ids1,"startyear":start_year, "endyear":end_year,"catalog":True, "calculations":True, "registrationkey":regkey})
p1 = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data1, headers=headers )

# Parse results.
j1 = json.loads(p1.text)

```

```python
# Check result.
# Parse results.
j1 = json.loads(p1.text)
dfs = []


series_count = 0


for serie in j1["Results"]['series']:
    serie_ID = str(serie["catalog"]["series_title"])
    data_json_list = []
    series_count += 1
    
    for data in serie['data']:
        
        data_json = {}
        data_json['year'] = data['year']
        data_json['period'] = data['period']
        data_json['periodName'] = data['periodName']
        data_json['date'] = data_json['period'][-2:]+str(data_json['year'])
        data_json[serie_ID] = data['value']
        
        data_json_list.append(data_json)
    
    
    
    if series_count == 1:

        df_final = pd.DataFrame(data_json_list)

        df_final['date']=pd.to_datetime(df_final['date'], format='%m%Y')

    else:
        
        df_series = pd.DataFrame(data_json_list)
        df_series['date']=pd.to_datetime(df_final['date'], format='%m%Y')

        sub_df = df_series[['date',serie_ID]]

        
        df_final = pd.merge(df_final,sub_df,how='outer',on='date')

#save all series into a  csv file
        
path = str('/Users/pedrobelotti/dropbox/berkeley/w209/df_final_jun_2022_IDs1title.csv') 

df_final.to_csv(path, index = False)
```
